* [0.2] List the main experimental(!) steps of the ChIP-seq protocol, its main advantages, and limitations.

→1. Сшивка: Этот этап включает обработку клеток сшивающим агентом для фиксации взаимодействий белок-ДНК.
2. Фрагментация хроматина: Затем сшитая ДНК расщепляется на мелкие фрагменты с помощью обработки ультразвуком или ферментативного расщепления.
3. Иммунопреципитация: Антитела, специфичные к интересующему белку, используются для разрушения комплексов белок-ДНК.
4. Очистка ДНК: ДНК выделяют из белковых комплексов и очищают для секвенирования.
5. Подготовка библиотеки: Очищенные фрагменты ДНК затем преобразуются в библиотеку секвенирования путем добавления адаптеров и амплификации фрагментов.
6. Секвенирование: Подготовленная библиотека секвенируется с использованием высокопроизводительных технологий секвенирования.

Основные преимущества ChIP-seq:
1. Профилирование по всему геному: ChIP-seq позволяет проводить картирование взаимодействий белок-ДНК по всему геному, обеспечивая всестороннее представление о сайтах связывания факторов транскрипции, модификациях гистонов и других белках, ассоциированных с хроматином.
2. Высокое разрешение: ChIP-seq может идентифицировать сайты связывания с высоким разрешением, что позволяет точно локализовать события связывания белка.
3. Чувствительность: ChIP-seq обладает высокой чувствительностью и может обнаруживать взаимодействия ДНК и белка с низким содержанием.

Ограничения ChIP-seq:
1. Смещение при сшивании: Этап сшивки может привести к смещениям при идентификации взаимодействий белок-ДНК.
2. Специфичность антител: Успех ChIP-seq зависит от наличия высококачественных антител, обладающих специфичностью к интересующему белку.
3. Задачи компьютерного анализа: Анализ данных ChIP-seq требует передовых инструментов биоинформатики и специальных знаний для точного определения сайтов связывания и интерпретации результатов.
4. Стоимость и время: ChIP-seq может быть дорогостоящим и отнимать много времени, особенно для крупномасштабных исследований или при работе с ограниченным количеством исходного материала.

→ 1. Лизис клеток и выделение ядер: Клетки подвергаются лизису с высвобождением ядер, которые затем выделяются из клеточного лизата.
2. Опосредованная транспозазой маркировка: Фермент транспозаза используется для одновременного фрагментирования и добавления адаптеров секвенирования к открытым участкам хроматина в изолированных ядрах.

5. Секвенирование: Подготовленная библиотека секвенируется с использованием высокопроизводительных технологий секвенирования.

Основные преимущества ATAC-seq:
1. Низкие требования к вводимым данным: ATAC-seq может быть выполнен всего с 500 клетками, что делает его подходящим для изучения редких популяций клеток или образцов с ограниченным исходным материалом.
2. Высокое разрешение: ATAC-seq может идентифицировать открытые участки хроматина с высоким разрешением, что позволяет точно отображать регуляторные элементы.
3. Скорость: ATAC-seq - относительно быстрый протокол, весь процесс от выборки до библиотеки секвенирования занимает всего несколько часов.

Ограничения ATAC-seq:
1. Смещение в сторону открытых областей хроматина: ATAC-seq может иметь смещение в сторону обнаружения открытых областей хроматина, потенциально отсутствующих регуляторных элементов в закрытых областях хроматина.
2. Доступность транспозазы: Эффективность опосредованной транспозазой маркировки может варьироваться в зависимости от различных областей генома, что приводит к потенциальным ошибкам при идентификации открытых областей хроматина.
3. Задачи вычислительного анализа: Подобно ChIP-seq, анализ данных ATAC-seq требует передовых инструментов биоинформатики и специальных знаний для точной идентификации открытых участков хроматина и интерпретации результатов.
4. Чувствительность к качеству пробы: ATAC-seq может быть чувствителен к качеству и целостности пробы, что требует тщательной оптимизации и мер контроля качества.

```python
# Возьмем из семинара функцию, подсчитывающая к-меры
from collections import defaultdict

# Inefficient but simple implementation
def calculate_kmers(seq: str, klen: int):
  assert len(seq) >= klen and klen >= 1, seq
  total_kmers = len(seq) - klen + 1

  counts = defaultdict(int)
  for ind in range(total_kmers):
    window = seq[ind:ind+klen]
    counts[window] += 1

  # Exclude non-ATGC k-mers
  counts = {
      k: v for k, v in counts.items() if {"A", "C", "G", "T"}.issuperset(set(k))
  }

  # Calculate frequencies
  total_kmers = sum(counts.values())
  frequencies = {k: v / total_kmers for k, v in counts.items()}
  return frequencies
```

```python
from tqdm import tqdm

KMERS = 1, 2, 3, 4, 5

# Список словарей (cls + features) - будущий DataFrame
df = []
for cls, sequences in (0, bgseq), (1, fgseq1), (2, fgseq2), (3, fgseq3):
  # tqdm рисует загрузку
  for seq in tqdm(sequences):
    record = {}
    for klen in KMERS:
      record.update(calculate_kmers(seq, klen))
    record['Class'] = cls
    df.append(record)
```

```python
# Мы будем делать на основе десишн три
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Класс случайный лес, методы фит и предикт
class CustomRandomForest:
    def __init__(self, n_estimators=100):
        self.n_estimators = n_estimators
        self.trees = [DecisionTreeClassifier() for _ in range(self.n_estimators)]

    def fit(self, X, y):
        for tree in self.trees:
            indices = np.random.choice(len(X), len(X), replace=True)
            tree.fit(X[indices], y[indices])

    def predict(self, X):
        predictions = np.zeros((len(X), self.n_estimators))
        for i, tree in enumerate(self.trees):
            predictions[:, i] = tree.predict(X)
        return np.mean(predictions, axis=1) >= 0.5

# Наши данные для задачи мультиклассификации, а так как задача состоит в том,
# чтобы написать простенький рандом форест, то я ограничусь простым собственным
# датасетом)
X = np.array([[1.0, 2.0], [3.0, 4.0]])
y = np.array([0, 1])

# Обучение и предсказание
custom_rf = CustomRandomForest()
custom_rf.fit(X, y)
custom_predictions = custom_rf.predict(X)

# Сравнение с склерн рандом форест
from sklearn.ensemble import RandomForestClassifier
sklearn_rf = RandomForestClassifier()
sklearn_rf.fit(X, y)
sklearn_predictions = sklearn_rf.predict(X)

print("Custom Random Forest Predictions:", custom_predictions)
print("Sklearn Random Forest Predictions:", sklearn_predictions)
```
